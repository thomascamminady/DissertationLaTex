```python
plt.style.use("kitishblank")
fig,ax = plt.subplots(1,1,figsize=(2,2))
x = np.array([0,np.cos(60/180*np.pi),1,0])
y = np.array([0,np.sin(60/180*np.pi),0,0])
plt.grid("off")
ax.set_xticks([])
ax.set_yticks([])
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
c = [kit.blue,kit.green,kit.blue]
matplotlib.rcParams['text.latex.preamble']=[r"\usepackage{amsmath}"]

for i in range(3):
    
    x0,y0,x1,y1 = x[i],y[i],x[i+1],y[i+1]
    ax.plot([x0,x1],[y0,y1],color=c[i],solid_capstyle='round',lw =4)
    xmid,ymid = (x0+x1)/2,(y0+y1)/2
    
    nx,ny = -(y1-y0),(x1-x0)
    ny *= 0.3
    nx *= 0.3
    
    ax.arrow(xmid,ymid,nx,ny,fc = 'k',head_width=0.05,)
    ax.text(xmid+nx*1.6,ymid+ny*1.6,r"$\boldsymbol{n}"+"_{}$".format(i+1),horizontalalignment="center",verticalalignment="center")


xmid,ymid = -0.2,-0.2
nx,ny = 0.2,0.1
xmid +=nx
ymid +=ny
nx*=-1
ny*=-1
ax.arrow(xmid,ymid,nx,ny,fc = 'k',head_width=0.05,)
posx = xmid
posy = ymid-0.12
ax.text(posx,posy,r"$\boldsymbol{\Omega}$",horizontalalignment="center",verticalalignment="center")

    
a = 0.8
center = 0.5
ax.set_xlim([center-a,center+a])
center = 0.2
ax.set_ylim([center-a,center+a])

plt.savefig("tri2pdf.pdf",constrained_layout=True)
```
