```python

import kitcolors as kit 
import matplotlib.pyplot as plt 
```

```python
plt.style.use("kitish")

fig, ax = plt.subplots()
off = 0.2

ax.spines['left'].set_position(('data', off))
#ax.spines['left'].set_color('none')
ax.set_ylabel("$y$",rotation = 0,labelpad = 10)
ax.set_yticks([])

ax.set_xticks([])
ax.set_xlabel("$x$")
ax.spines['bottom'].set_position(('data', off))


ax.arrow(off,off,0.7,0.3,head_width=0.04,linewidth = 0.6, color = "k")
ax.text(0.8,0.5,"$\mathbf{\Omega}$")#,fontsize = 16)
#ax.spines['left'].set_position('center')
#ax.spines['bottom'].set_position('center')

ax.axvline(off+0.7,off,off+0.3,lw = 1, color = "k", linestyle = "--")
ax.axhline(off,0.1,0.45,color = kit.green,zorder = 100,lw = 3,solid_capstyle='round')
ax.text(0.6,0.25,"$\mu$",color = kit.green)#,fontsize = 20)
ax.set_xlim([0,2])
ax.set_ylim([0,1])
plt.savefig("slab.pdf")
```
