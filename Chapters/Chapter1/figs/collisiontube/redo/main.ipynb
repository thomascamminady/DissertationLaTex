```python
import matplotlib.pyplot as plt
#import seaborn as sns
#import seaborn as sns
import kitcolors as kc
#sns.set_context("poster")
```

```python
plt.style.use("kitish")
fig, ax = plt.subplots(1,1)#,figsize=(20,6))
ax.set_xticks([])
ax.set_yticks([])
ax.set_xticklabels([])
ax.set_yticklabels([])
fig.patch.set_visible(False)
plt.tight_layout()
ax.set_facecolor(kc.green)
for border in ["bottom","top","right","left"]:
    ax.spines[border].set_color(kc.green)
r = 0.1

ax.set_xlim([0,2])
ax.set_ylim([0,0.6])
x = [0.2,0.35,0.5,1.2,1.5,1.6,1.8,0.2-r]
y = [0.3,0.06,0.55,0.081,0.2,0.5,0.6,0.3-2*r+r/4]
cs= ["w","k","k","k",kc.orange,"k","k","k"]
for i,(xi,yi,c) in enumerate(zip(x,y,cs)):
    circle1 = plt.Circle((xi,yi), r, color=c)
    ax.add_artist(circle1)
    circle1 = plt.Circle((xi,yi), 0.005, color="w")
    #plt.text(xi-0.05,yi-0.03,r"$(x_{},y_{})$".format(i,i),color="w",fontsize = 20)
    ax.add_artist(circle1)
plt.text(x[4]-0.08,y[4]-0.05,r"$(x_i,y_i)$",color="w")
   

x0,y0,L= x[0],y[0],1.3
ax.plot([x0,x0+L],[y0-r,y0-r],"w--")
ax.plot([x0,x0+L],[y0+r,y0+r],"w--")
ax.plot([x0+L,x0+L],[0.2,y0+r],"w--")

#ax.plot([x0,x0+L],[y0+r+0.02,y0+r+0.02],"w-")

plt.arrow(x0,y0,0.2,0,color="w",width = 0.01)
plt.text(x0-0.1+0.005,y0-0.04,r"$(x_{n},y_{n})$",color="k",ha = "left")
#plt.text(x0+0.23,y0-0.03,r"$(v_x,v_y)$".format(0,0),color="k",fontsize = 20)
plt.text(1.0,y0+r+0.02,r"$\delta t_n$",color = "w")#,fontsize = 50)

ax.plot([x0,0.98],[y0+r+0.03,y0+r+0.03],c="w")
ax.plot([1.13,x0+L],[y0+r+0.03,y0+r+0.03],c="w")

ax.plot([x0,x0],[y0+r+0.03-0.01,y0+r+0.03+0.01],c="w")
ax.plot([x0,x0],[y0+r+0.03-0.01,y0+r+0.03+0.01],c="w")
ax.plot([x0+L,x0+L],[y0+r+0.03-0.01,y0+r+0.03+0.01],c="w")
ax.plot([x0+L,x0+L],[y0+r+0.03-0.01,y0+r+0.03+0.01],c="w")


circle1 = plt.Circle((x0,y0), 0.005, color="k")
ax.add_artist(circle1)

plt.tight_layout()
plt.savefig("cylinder.pdf")
```
