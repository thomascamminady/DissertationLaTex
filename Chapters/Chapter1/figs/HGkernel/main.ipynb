```python
import numpy as np
import matplotlib.pyplot as plt
#import seaborn as sns
#sns.set_context("poster")
import matplotlib as mpl
#from camminapy import * 
```

```python
def s(g,mu):
    return (1-g**2)/4/np.pi/(1-2*g*mu+g**2)**1.5
```

```python
theta = np.linspace(0,2*np.pi,10000);
mu = np.cos(theta)
```

```python
plt.style.use("kit2")
#mpl.rcParams['mathtext.fontset'] = 'cm'
#mpl.rcParams['mathtext.rm'] = 'serif'
fig = plt.figure(figsize=(6,6))
#ax = plt.subplot(111,projection ='polar')
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8],
                  projection='polar')


gs = [0.0,0.7,0.9,0.99][::-1]


#kitcolors = getkitcolors()
#colors = [kitcolors["green"],kitcolors["blue"],kitcolors["orange"],kitcolors["red"],kitcolors["cyan"],kitcolors["maygreen"]]
for g in gs:
    z = s(g,np.cos(theta))
    ax.plot(theta,np.log10(z))
#ax.grid(True)

#plt.thetagrids([0,90,180,270])
ax.set_theta_zero_location("N")
ax.set_rticks([])
ax.legend([r"$g={}$".format(g) for g in gs])
ax.set_rlabel_position(200)
#ax.set_title(r"Henyey-Greenstein kernel $\log_{10}\,s_g(\cos\, \theta )$",pad = 5)
# tick locations
thetaticks = np.arange(0,310,45)
plt.tick_params(pad = 15)
# set ticklabels location at 1.3 times the axes' radius
ax.set_thetagrids(thetaticks,fontsize = 20)#,labelpad = 10)#, frac=1.3)
plt.tight_layout()
plt.tight_layout(rect=[0, 0.03, 1, 1.1])
plt.savefig("henyeygreensteinpolar.png")
plt.savefig("henyeygreensteinpolar.pdf")
```
