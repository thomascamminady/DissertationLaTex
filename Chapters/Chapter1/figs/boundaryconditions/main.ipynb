```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_context("poster")
import numpy as np
from numpy import linspace
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import to_rgba_array
```

```python
t = np.linspace(0,1,1000)
x = 7*np.cos(2*np.pi*t)
y = (4*np.sin(2*np.pi*t)+ 0.4*np.cos(4*np.pi*t)+ 5*np.exp(-100*(t-0.2)**2) -1.25)/4.5
```

```python
N = 255
nalphas = 4
alphas = linspace(1,0.2,nalphas)
#alphas = [1,0.8,0.6,0.4,0.2] # choose as many alphas as you would like to have 

kitgreen =(0/N,150/N,130/N)
kitblue  =(70/N,100/N,170/N)
kitmaygreen=(140/N,182/N,60/N)
kityellow =(252/N,229/N,0/N)
kitorange=(233/N,155/N,27/N)
kitbrown = (167/N,130/N,46/N)
kitred =(162/N,34/N,35/N)
kitlila = (163/N,16/N,124/N)
kitcyan = (35/N,161/N,224/N)
colors=[kitgreen,kitorange,kitblue,kitlila,kitmaygreen,kitbrown,kityellow,kitcyan,kitred]
kitcolors = []
```

```python
kitgreen = (0,0,0)
plt.style.use("kit2")
fig, ax = plt.subplots(1,1,figsize=(20,5))
ax.plot(x,y,color =kitgreen)
plt.axis('off')
plt.savefig("domain_with_boundary.png", bbox_inches='tight',dpi = 400)
ax.text(0,0,r"$V$",fontsize = 40)
ax.text(-4,0.95,r"$\partial V$",fontsize = 40,color = kitgreen)
k = 100
dx =-(x[k+1] -x[k])
dy =(y[k+1] - y[k])
norm = (dx**2+dy**2)**0.5
dx /= norm
dy /= norm
print(dx)
print(dy)
ax.set_xlim([-8,8])
ax.set_ylim([-2,2])
ax.text(6.5,1.2,r"${\bf{n}}$",fontsize=40)
plt.arrow(x[k],y[k],0.5*dx,0.5*4*dy,head_width=0.1,ec ="k",fc="k")
plt.savefig("domain_with_boundary.png", bbox_inches='tight',dpi = 400)
```
