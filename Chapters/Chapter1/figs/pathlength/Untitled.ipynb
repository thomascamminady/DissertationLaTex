```python
import numpy as np
import matplotlib.pyplot as plt
import scipy 
import matplotlib
import kitcolors as kit
```

```python
x = np.loadtxt("file.txt")
```

```python
len(x)
```

```python
plt.style.use("kitish")
fig,ax = plt.subplots(1,1)

hist,edges = np.histogram(x,bins = 30,density=True)
centers = (edges[:-1] + edges[1:])/2
ax.scatter(centers,hist,label="Data",s=40)
c = 4.007
ax.plot(centers,c*np.exp(-c*centers),linestyle="-",label=r"Best fit $\lambda \, \exp(-\lambda s),\quad  \lambda = {}$".format(c),color = kit.black,zorder = -1)
plt.yscale("log")
#ax.set_ylim([1,10**6])
ax.set_xlim([0,3.5])
ax.set_xlabel(r"Distance $s$")
ax.set_ylabel("Frequency")
#ax.set_title("Path-length histogram for $n=10^6$ samples.")
plt.legend()
plt.savefig("kithistogram.png")
plt.savefig("kithistogram.pdf")
```
