```python
import numpy as np
import matplotlib.pyplot as plt
import os 
import kitcolors as kit
```

```python
plt.style.use("kitish")
fig, ax = plt.subplots(1,1)
x = np.array([-10,10])
m = 0.1
ax.plot(x,x*m,"--",color = kit.black)
x = np.array([-8,8])
m = 0.1
ax.plot(x,x*m,color = kit.black)
x0 = -7
dx = -2
A = 0.25
ax.arrow(x0,A+0,dx,m*dx,width = 0.01,head_width=0.1,facecolor="k",head_length=0.1)
ax.text(-9,A-0.1,r"$z$",horizontalalignment="left")

ax.arrow(x0,A+0,0,-3*m*dx,width = 0.02,head_width=0.3,facecolor="k",head_length=0.05)
ax.text(-6.7,A+0.6,r"$x$",horizontalalignment="left")

ax.arrow(x0,0+A,30/3.1*m*dx,-3/3.1*m*dx,width = 0.01,head_width=0.09,facecolor="k",head_length=0.1)
ax.text(-8.8,A+0.3,r"$y$",horizontalalignment="left")

ax.scatter(x0,0+A,s=10,c="k")

ax.set_xticks([])
ax.set_yticks([])
ax.set_axis_off()


X = np.linspace(-10,10,16)[2:-2]
ax.scatter(X,m*X,c=[kit.green]*len(X),s =50,zorder = 100,alpha=1)#,zorder = 10)
ax.scatter(X,m*X,c=[kit.green]*len(X),s =1000,alpha = 0.5,edgecolor=kit.green,zorder=10)#,zorder = 10)
   

filename ="sketch.pdf"
fig.savefig(filename)
os.system("pdfcrop --margins '1 1 1 1 ' {} {}".format(filename,filename))    
```
