```python
import numpy as np
import pandas as pd 
import os
import numpy as np 
from camminapy import snpp
```

```python
prefix = "/home/qd4314/Work/Projects/rSN/code/out/"
sf = [x for x in os.listdir(prefix)]
df = pd.DataFrame()
print(len(sf))
for i, folder in enumerate(sf):
    print(i,end="")
    foldername = prefix + "/" + folder
    try:
        config = snpp(foldername)
        df = df.append(config, ignore_index=True)
        print("!",end=",")
    except:
        pass
df.to_csv("iso.csv")
```
