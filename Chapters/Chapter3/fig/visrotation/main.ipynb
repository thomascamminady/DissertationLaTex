```julia
using Revise
include("newquadrature/Quadrature.jl")
using DelimitedFiles
using LinearAlgebra
println("done")
```

```julia
order = 10
qtype = 2

Q = Quadrature(order,qtype,false)
prefix = "l"
qtypes = string(qtype)
writedlm(qtypes*prefix*"points.txt",Q.pointsxyz)
writedlm(qtypes*prefix*"weights.txt",Q.weights)
writedlm(qtypes*prefix*"neighbours.txt",Q.neighbors)
writedlm(qtypes*prefix*"triangles.txt",Q.triangulation)

Q2 = Quadrature(order,qtype,true)
prefix = "s"
writedlm(qtypes*prefix*"points.txt",Q2.pointsxyz)
writedlm(qtypes*prefix*"weights.txt",Q2.weights)
writedlm(qtypes*prefix*"neighbours.txt",Q2.neighbors)
writedlm(qtypes*prefix*"triangles.txt",Q2.triangulation)
println("done")
```
