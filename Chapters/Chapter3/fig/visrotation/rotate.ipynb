```julia
using Revise
include("newquadrature/Quadrature.jl")
using DelimitedFiles
using LinearAlgebra
println("done")
```

```julia
order = 15
qtype = 2
Q = Quadrature(order,qtype,false)

prefix = "l"
qtypes = string(qtype)
writedlm("rotate_points.txt",Q.pointsxyz)
writedlm("rotate_weights.txt",Q.weights)
writedlm("rotate_neighbours.txt",Q.neighbors)
writedlm("rotate_triangles.txt",Q.triangulation)
```

```julia
nq = length(Q.weights)
phi = zeros(nq,1,1)
phinew = zeros(nq,1,1)
x0 = [0.0,-1.0,0.0]
sigma = 0.2
for i=1:nq
    phi[i,1,1] =exp(-norm(Q.pointsxyz[i,:]-x0)^2/(2*sigma^2)) /sqrt(2*pi*sigma^2)
    xyz = Q.pointsxyz[i,:]
    if xyz[1]<=0.0001 && xyz[2] <= 0.00001 && xyz[3]>=-0.001
        phi[i,1,1] = 1.0
    else
        phi[i,1,1] = 0.0
    end
end
print(nq)
writedlm("values_0.txt",phi)
axis = [0.0,0.0,1.0]
theta = 0.1
for i=1:20
    x = randn(3)
    axis = x/norm(x)
    rotateandinterpolate!(axis,theta,Q,phinew,phi,"forward")
    phi = deepcopy(phinew)
    phinew = zeros(nq,1,1)
    rotateandinterpolate!(axis,-theta,Q,phinew,phi,"forward")
    phi = deepcopy(phinew)
    phinew = zeros(nq,1,1)
    if i in [5,10,15,20]
    writedlm(string(i)*"values.txt",phi[:,1,1])
    end


end
```
