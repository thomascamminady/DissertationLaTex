```python
import numpy as np
import matplotlib.pyplot as plt
import os 
```

```python
path = "./"
directories = os.listdir( path )
dirs = [x  for x in directories if x[:2] == "20"]
print(dirs) 
```

```python
n = len(dirs)

plt.style.use("kitishnotex")
fig,axs = plt.subplots(n,1,figsize=(2,n*2))
data = []
for i,direc in enumerate(dirs):
    d = np.loadtxt(direc+"/data/rhofinal.txt",delimiter = ",")
    data.append(d)
    axs[i].pcolormesh(d)
```

```python
d1 = data[0]
d2 = data[1]
```

```python
im = plt.pcolormesh((d1-d2)/(1e-16+d1))
plt.colorbar(im)
```
