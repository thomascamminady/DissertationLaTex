```python
import plotly.express as px
import numpy as np
import pandas as pd 
```

```python
qp = np.loadtxt("data/quadpoints.txt",delimiter = ",")
```

```python
qw = np.loadtxt("data/quadweights.txt")
```

```python
df = pd.DataFrame({"x":qp[:,0],
                   "y":qp[:,1],
                   "z":qp[:,2],
                   "w":qw})
fig = px.scatter_3d( df,x="x",y="y",z="z",color="w")
fig.show()
```
