```python
import matplotlib.pyplot as plt
import kitcolors as kit
import numpy as np
from matplotlib.patches import Circle
from matplotlib.collections import PatchCollection
```

```python
# Situation 1
n = 8
sigma = 1
r = sigma/n**2

pl = np.random.rand(n**2,2)
pl[:,0] *= 0.95

xr,yr = np.meshgrid(np.linspace(1,2,n),np.linspace(0,1,n))
pr = np.vstack([xr.flatten(),yr.flatten()]).T

plt.style.use("kitish")
scale = 2
fig,ax = plt.subplots(figsize=(2*scale,scale))
patches = []
for i in range(n**2):
    ax.add_artist(Circle((pl[i,0],pl[i,1]),r,color = kit.green))

patches = []
for i in range(n**2):
    ax.add_artist(Circle((pr[i,0],pr[i,1]),r,color = kit.blue))



ax.set_xlim([0-2*r,2+2*r])
ax.set_ylim([0-2*r,1+2*r])
ax.set_xticks([])
ax.set_yticks([])
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)

plt.savefig("truehet.pdf")
```

```python
# Situation 1
n = 8
sigma = 1
r = sigma/n**2

xr,yr = np.meshgrid(np.linspace(0,1,n//2),np.linspace(0,1,n//2))
pl = np.vstack([xr.flatten(),yr.flatten()]).T
pl *= 0.8
pl += 0.1


xr,yr = np.meshgrid(np.linspace(1,2,n),np.linspace(0,1,n))
pr = np.vstack([xr.flatten(),yr.flatten()]).T

plt.style.use("kitish")
scale = 2
fig,ax = plt.subplots(figsize=(2*scale,scale))
patches = []
for i in range(n**2//4):
    ax.add_artist(Circle((pl[i,0],pl[i,1]),4*r,color = kit.green))

patches = []
for i in range(n**2):
    ax.add_artist(Circle((pr[i,0],pr[i,1]),r,color = kit.blue))



ax.set_xlim([0-2*r,2+2*r])
ax.set_ylim([0-2*r,1+2*r])
ax.set_xticks([])
ax.set_yticks([])
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)

plt.savefig("falsehet.pdf")
```
