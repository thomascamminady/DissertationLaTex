```python
import kitcolors as kit
import matplotlib.pyplot as plt 
import matplotlib.patches as patches
```

```python
plt.style.use("kitish")
fig, ax = plt.subplots(figsize=(5,1.5*1.875))
ax.set_xlim(-2,2)
ax.set_ylim(-1,1)

styledict = dict(facecolor=kit.black10, edgecolor='black', boxstyle='round,pad=1')
ax.text(-0.75, 0.25, 'Correlated\nParticles', color='black', bbox=styledict,ha="center",va="center")
ax.text(0.75, 0.25, 'Uncorrelated\nParticles', color='black', bbox=styledict,ha="center",va="center")


styledict = dict(facecolor='none', edgecolor='black', boxstyle='round,pad=1')
ax.text(-0.75, -0.75, 'Correlated\nSource', color='black', bbox=styledict,ha="center",va="center")
ax.text(0.75, -0.75, 'Uncorrelated\nSource', color='black', bbox=styledict,ha="center",va="center")



style = "Simple, tail_width=0.5, head_width=4, head_length=8"
kw = dict(arrowstyle=style, color="k")

arr = patches.FancyArrowPatch((0.6, 0.6), (-0.6, 0.6),connectionstyle="arc3,rad=.5", **kw)
ax.add_patch(arr)
ax.text(0,1.0,"Scattering",va = "center",ha="center")
arr = patches.FancyArrowPatch((0.75, -0.5), (0.75, -0.),connectionstyle="arc3", **kw)
ax.add_patch(arr)
ax.text(-0.8,-0.4,"Emission",ha="right")
arr = patches.FancyArrowPatch((-0.75, -0.5), (-0.75, -0.),connectionstyle="arc3", **kw)
ax.add_patch(arr)
ax.text(0.8,-0.4,"Emission",ha="left")


arr = patches.FancyArrowPatch((-0.8,0.5), (-1.1,0.5),connectionstyle="arc3,rad=3", **kw)
ax.add_patch(arr)
ax.text(-1.1,0.75,"Scattering",ha="right")


arr = patches.FancyArrowPatch((-1.25, 0.25), (-1.85, 0.25),connectionstyle="arc3", **kw)
ax.text(-1.6,0.35,"Absorption",ha="right")
ax.add_patch(arr)
arr = patches.FancyArrowPatch((1.25, 0.25), (1.85, 0.25),connectionstyle="arc3", **kw)
ax.text(1.6,0.35,"Absorption",ha="left")

ax.add_patch(arr)


plt.grid("off")
ax.set_xticks([])
ax.set_yticks([])
ax.spines['left'].set_color('none')
ax.spines['bottom'].set_color('none')
plt.savefig("flowchart1.pdf")


kw = dict(arrowstyle=style, color=kit.red)

arr = patches.FancyArrowPatch((-0.6, 0), (0.6, 0),connectionstyle="arc3,rad=.5", **kw)
ax.add_patch(arr)
ax.text(0,-0.4,"Crossing\nInterface",va="center",ha="center",color =kit.red)
plt.savefig("flowchart2.pdf")
```
