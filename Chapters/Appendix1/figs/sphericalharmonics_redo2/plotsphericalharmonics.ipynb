```python
from scipy.special import sph_harm as Y 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import BoundaryNorm
from matplotlib.ticker import MaxNLocator
import kitcolors as kit
from camminapy import * 
from matplotlib.colors import LinearSegmentedColormap
n_bin = 101
import colorcet as cc
```

```python
ntheta,nphi = 200,100
theta,phi = np.linspace(0,2*np.pi,ntheta),np.linspace(0,np.pi,nphi)
theta,phi = np.meshgrid(theta,phi) 
```

```python
nmax = 4
scale = 0.6
plt.style.use("kitish")
fig, axs = plt.subplots(1*nmax,1*nmax,
                        figsize=(11.69,8.27)) # 
                        #figsize=(20*nmax*scale,15*nmax*scale))
for n in range(nmax):
    for m in range(n+1,nmax):
        axs[n,m].axis('off')
    for m in range(n+1):
        ax = axs[n,m]
        z = np.real(Y(m,n,theta,phi))
        # Colorbar
        levels = MaxNLocator(nbins=201).tick_values(-0.8,0.8)#z.min(), z.max())
        cmap = mycmapdiv(kit.cyan,"white",kit.red,nbins = 201)
        
#        cmap = LinearSegmentedColormap.from_list("mystyle", [kit.blue,"white",kit.yellow], N=201)
   #     cmap = cc.cm.CET_D10#coolwarm#diverging_bwr_55_95_c37#40_95_c42

    #
        norm = BoundaryNorm(levels, ncolors=cmap.N, clip=True)

        im = ax.pcolormesh(z,cmap=cmap,norm = norm , linewidth=0,rasterized=True)
        if True:
            if n==0 and m==0 :
                cbaxes = fig.add_axes([0.3, 0.9, 0.6, 0.04]) 
                #cb = plt.colorbar(ax1, cax = cbaxes)  
                cbar = fig.colorbar(im, ax=axs[0,2],ticks = [-0.8,0,0.8],cax = cbaxes,orientation ="horizontal")#fraction = 0.15,shrink = 2,orientation ="horizontal")#z.min(),z.max()])

            #else:
            #    mid = (z.min()+z.max())/2
            #    cbar = fig.colorbar(im, ax=ax,ticks = [np.round(z.min(),3),np.round(mid,3),np.round(z.max(),3)])

        # Labels
        if n==3:
            ax.set_xticks([0,50,100,150,200])
            ax.set_xticklabels([r"$0$",r"$\pi/2$", r"$\pi$",r"$3\pi/2$",r"$2\pi$"])
            ax.set_xlabel(r"$\theta$")
        else:
            ax.set_xticks([0,50,100,150,200])
            ax.set_xticklabels(["","","","",""])

        if m==0:
            ax.set_yticks([0,100])
            ax.set_yticklabels([r"$0$",r"$\pi$"])
            ax.set_ylabel(r"$\phi$")#,rotation = 0,labelpad = 6)
           # ax.yaxis.set_label_coords(-0.15, 0.7)
        else:
            ax.set_yticks([0,50,100])
            ax.set_yticklabels(["","",""])
        ax.set_title(r'Re$\left(Y_{}^{}(\theta,\phi)\right)$'.format(n,m), pad= 5)
plt.tight_layout()
plt.savefig("REAL.pdf",dpi = 400)
```
