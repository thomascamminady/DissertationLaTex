```python
import numpy as np
import kitcolors as kit
import matplotlib
import matplotlib.pyplot as plt 
import matplotlib.collections
import matplotlib.transforms as mtransforms
import matplotlib.patches as mpatch
from matplotlib.patches import FancyBboxPatch
```

```python
plt.style.use("kitish")
w,h = 5, 1.875/2
```

```python
factor = 1
fig, ax = plt.subplots(figsize=(w,h))



np.random.seed(3)
n  = 300
x,y = w/2*np.random.rand(n), h*np.random.rand(n)
s = 0.02
patches = [plt.Circle((xi,yi), s) for xi,yi in zip(x,y)]
coll = matplotlib.collections.PatchCollection(patches, facecolors=kit.green)
ax.add_collection(coll)

step = 0.1

x = np.arange(w/2,w,step = step)
y = np.arange(0,h,step = step)
x,y = np.meshgrid(x,y)
x,y = x.flatten(), y.flatten()

s = 0.02


# Note that the patches won't be added to the axes, instead a collection will
patches = [plt.Circle((xi,yi), s) for xi,yi in zip(x,y)]

coll = matplotlib.collections.PatchCollection(patches, facecolors=kit.blue)
ax.add_collection(coll)







plt.axis('off')
ax.set_xlim([0,w])
ax.set_ylim([0,factor*h])

plt.savefig("truehet.pdf")
```
