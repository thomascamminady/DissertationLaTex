```python
import numpy as np
import matplotlib.pyplot as plt
import kitcolors as kit
```

```python
x = np.linspace(0,10,1000)
l1,l2,l3 = 1,2,3
y1 = np.sin(l1*np.pi*x)*np.exp(-x/l1)+l1/(1+x/2)
y2 = np.sin(l2*np.pi*x)*np.exp(-x/l2)+l2/(1+x/5)
y3 = np.sin(l3*np.pi*x)*np.exp(-x/l3)+l3/(1+x/10)
```

```python

plt.style.use("kitishnotex")
fig, axs = plt.subplots(3,2,figsize =(10,6),facecolor='w')

ax = axs[0,0]
ax.plot(x,y1,label = "$y_1(x)$")
ax.plot(x,y2,label = "$y_2(x)$")
ax.plot(x,y3,label = "$y_3(x)$")
ax.legend(loc = "upper right",ncol = 3)
ax.set_xlabel("$x$")
ax.set_title("Version 1A: everything in the legend")



ax = axs[1,0]
ax.plot(x,y1,label = "$i=1$")
ax.plot(x,y2,label = "$i=2$")
ax.plot(x,y3,label = "$i=3$")
ax.legend(loc = "upper right",ncol = 3)
ax.set_xlabel("$x$")
ax.set_ylabel("$y_i(x)$")
ax.set_title("Version 2A: y-label + legend")



ax = axs[2,0]
ax.plot(x,y1,label = "$i=1$")
ax.plot(x,y2,label = "$i=2$")
ax.plot(x,y3,label = "$i=3$")
ax.text(x[-1]+0.2,y1[-1], "$y_1(x)$", va = "center", color = kit.green)
ax.text(x[-1]+0.2,y2[-1], "$y_2(x)$", va = "center", color = kit.black)
ax.text(x[-1]+0.2,y3[-1], "$y_3(x)$", va = "center", color = kit.blue)
ax.set_xlabel("$x$")
ax.set_title("Version 3A: text at (end of) line")





ax = axs[0,1]
ax.plot(x,y1,label = "$y_1(x)$")
ax.legend(loc = "upper right")
ax.set_xlabel("$x$")
ax.set_title("Version 1B: everything in the legend")



ax = axs[1,1]
ax.plot(x,y1,label = "$i=1$")
ax.legend(loc = "upper right")
ax.set_xlabel("$x$")
ax.set_ylabel("$y_i(x)$")
ax.set_title("Version 2B: y-label + legend")



ax = axs[2,1]
ax.plot(x,y1,label = "$i=1$")
ax.text(x[-1]+0.2,y1[-1], "$y_1(x)$", va = "center", color = kit.green)
ax.set_xlabel("$x$")
ax.set_title("Version 3B: text at (end of) line")


plt.suptitle("Ways to label graphs (many lines or single line). Assume $y_i(x)$ is unit-less. Choose two.")

plt.savefig("labels.png")
```
